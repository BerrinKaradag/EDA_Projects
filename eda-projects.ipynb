```python
df.columns
```

```python
#deck değikenindeki boş değerleri deck değişkenin en çok tekrar eden değeri (mode) ile doldur:

a=df['deck'].mode()[0] # en sık tekrar eden dataların olduğu list verir, [0] diyelerek ilk veriyi alırsın
print(a)
```

```python
df["deck"]=df["deck"].fillna(a) 
df.head()

#.fillna null değerleri doldur demek, inplace=true df içinde yap demek
```

```python
# age değikenindeki boş değerleri age değişkenin medyanı ile doldur:

b=df["age"].median()
df["age"]=df["age"].fillna(b)
```

```python
df.groupby(["pclass","sex"])["survived"].agg(["sum","count","mean"]).reset_index()
```

30 yaşın altında olanlar 1, 30'a eşit ve üstünde olanlara 0 vericek bir fonksiyon yazın. Yazdığınız fonksiyonu kullanarak titanik veri
setinde age_flag adında bir değişken oluşturunuz oluşturunuz.

```python
def fonk (x):
    if x<30:
        return 1
    else:
        return 0
df["age_flag"]=df["age"].apply(fonk)    


# df["age_flag"]=df["age"].apply(lambda x: fonk(x))

# df["age_flag"]=df["age"].apply(lambda x: 1 if x<30 else 0)

df.head()
```

# "tips" Dataset EDA

```python
x=sns.load_dataset("tips")
x.head()
```

```python
x.groupby("time",observed=True)["total_bill"].agg(["min","max","mean"]).reset_index() 
```

```python
x.groupby(["day","time"],observed=True)["total_bill"].agg(["min","max","mean"]).reset_index()
```

Lunch zamanına ve kadın müşterilere ait total_bill ve tip değerlerinin day'e göre toplamını, min, max ve ortalamasını bulunuz.

```python
x.columns
x[(x["sex"]=="Female") & (x["time"]=="Lunch")].groupby("day",observed=True)[["total_bill","tip"]].agg(["min","max","mean"]).reset_index()
```

```python
z = x.loc[(x["size"] < 3) & (x["total_bill"] > 10) , "total_bill"].mean()

print(z)
```

```python
x["total_bill_tip_sum"]=x["total_bill"]+x["tip"]
x.head()
```

```python
q=x.sort_values(by="total_bill_tip_sum", ascending=False)
q.head(30)
```
